```python
#  ПУНКТ 2 ISSUE 
# 1. Что такое float — простыми словами

# Тип данных float (числа с плавающей точкой)
# float — это вещественное число, то есть число с дробной частью

# Примеры вещественных чисел:

        # 3.14

        # 0.5

        # −2.75

        # В Python такие числа имеют тип float.

#2. Почему тип называется float

#float — сокращение от floating point
#(по-русски: число с плавающей точкой).

#Что значит «плавающая точка»

# Это означает, что:

    # точка не фиксирована

    # число хранится в виде: мантисса × 10^степень 
    # Пример:
            # 314 × 10⁻² = 3.14
            # 1 × 10³ = 1000
# 3. Самый важный факт (его надо принять)
# float — НЕ точный тип
# Это не особенность Python

# 4. Почему float неточен (досконально)
    # Компьютер хранит числа: не в десятичной системе (0–9),
    # а в двоичной (0 и 1).
    # Проблема
    # Некоторые десятичные дроби невозможно точно представить в двоичной системе.
    # Пример: 1/10 (0.1) — в двоичной системе это бесконечная дробь
    # Поэтому Python хранит приближение, а не точное значение.


# 5. Классический пример неточности
first_float = 0.1
second_float = 0.2
float_sum = first_float + second_float
print(float_sum)

# Математически ты ожидаешь: 0.3 
# Но на самом деле: 0.30000000000000004
    # вывод (запомнить)
    # float → приближённые вычисления
    # float ❌ нельзя использовать:для денег для строгого сравнения ==
    # float ✅ используют:в физике в графике в измерениях в науке


# 7. Как создаётся float
pi_float = 3.14
ten_as_float = 10.0

# Способ 2. Конструктор float()
parsed_float = float("3.14")
integer_string_as_float = float("10")
# Результаты:

            # "3.14" → 3.14

            # "10" → 10.0

# 8. Что float() умеет парсить
                                # float("3.14")
                                # float(" 3.14 ")
                                # float("-2.5")
                                # float("+1.0")

# Научная нотация (очень неочевидно) 1 × 10³ = 1000.0
scientific_float = float("1e3")
print(scientific_float)

# Не  работает
                # float("3,14")  # ValueError (запятая — не формат числа)
                # float("abc")   # ValueError
                # float("")      # ValueError

# Отличие float("3.0") от int("3.0")  
                                    # float("3.0")  
                                    # # 3.0 int("3.0") ValueError  
                                    # Почему
                                            # float понимает дробный формат
                                            # int — только целые числа


# 9. Сравнение float — ловушка

# Плохо result_float == 0.3 
# Правильно (в инженерных задачах): abs(result_float - 0.3) < 1e-9


# 10. Особые значения float (знать, но не использовать)  

# inf — бесконечность
positive_infinity = float("inf") 


negative_infinity = float("-inf")

# nan — “не число”
not_a_number = float("nan")
print(positive_infinity)
print(negative_infinity)
print(not_a_number)

# Пример: создание числа с плавающей точкой из строки
float_from_string = float("3.14")

# Пример: сложение float
first_float = 0.1
second_float = 0.2
float_sum = first_float + second_float
print(float_sum)
# float_sum не равно ровно 0.3 из-за представления float
 

# 15. Короткая выжимка (как для собеса)
# float — приближённый тип
# хранится в двоичной системе
# неточен по природе
# float("3.0") работает, int("3.0") — нет
# в парсинге идёт после int
```
